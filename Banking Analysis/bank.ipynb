```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("bank.csv")
```

```python
df.columns
```

```python
x=df.iloc[ : , :-1]
x.head()
```

```python
y=df.iloc[ : ,-1]
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(xtrain,ytrain)
ypred=knn.predict(xtest)
```

```python
accuracy_score(ytest,ypred)
```

```python
print(classification_report(ytest,ypred))
```

```python
print(confusion_matrix(ytest,ypred))
```

###### Hyper-parameter tuning

```python
accuracy=[]
for i in range(1,21):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    ypred=knn.predict(xtest)
    ac=accuracy_score(ytest,ypred)
    accuracy.append(ac)
```

```python
plt.plot(range(1,21),accuracy)
plt.grid(True)
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=15)
knn.fit(xtrain,ytrain)
ypred=knn.predict(xtest)
```

```python
accuracy_score(ytest,ypred)
```

```python
print(classification_report(ytest,ypred))
```

```python
print(confusion_matrix(ytest,ypred))
```

###### Cross Validation Score

```python
from sklearn.model_selection import cross_val_score
cvs=cross_val_score(knn,x,y,cv=5,scoring="accuracy")
print(f"CVS: {cvs}")
print(f"CVS Mean: {cvs.mean()}")
print(f"CVS Standard Deviation: {cvs.std()}")
```

###### Criterion---> "Gini"

```python
s=0
for i in range(1,50):
    dt2=DecisionTreeClassifier(max_depth=i)
    dt2.fit(xtrain,ytrain)
    ypred=dt2.predict(xtest)
    ss=accuracy_score(ytest,ypred)
    if(s<ss):
        s=ss
        a=i
print(s,a)
```

```python
s=0
for i in range(1,50):
    dt3=DecisionTreeClassifier(min_samples_leaf=i)
    dt3.fit(xtrain,ytrain)
    ypred=dt3.predict(xtest)
    ss=accuracy_score(ytest,ypred)
    if(s<ss):
        s=ss
        a=i
print(s,a)
```

###### Criterion--->"Entropy"

```python
s=0
for i in range(1,50):
    dt4=DecisionTreeClassifier(criterion="entropy",max_depth=i)
    dt4.fit(xtrain,ytrain)
    ypred=dt4.predict(xtest)
    ss=accuracy_score(ytest,ypred)
    if(s<ss):
        s=ss
        a=i
print(s,a)
```

```python
s=0
for i in range(1,50):
    dt5=DecisionTreeClassifier(criterion="entropy",min_samples_leaf=i)
    dt5.fit(xtrain,ytrain)
    ypred=dt5.predict(xtest)
    ss=accuracy_score(ytest,ypred)
    if(s<ss):
        s=ss
        a=i
print(s,a)
```
